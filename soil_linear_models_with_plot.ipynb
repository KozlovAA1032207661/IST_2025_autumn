# Линейные модели для температуры почвы — финальная версия с графиком предсказаний

В ноутбук добавлена визуализация:
- реальные значения температуры почвы
- предсказания линейной регрессии
- график строится на контрольной выборке


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, KFold, GridSearchCV, cross_val_predict
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# ===== ЗАГРУЗКА ДАННЫХ =====
df = pd.read_excel('rp5_with_soil_ml_scaled.xlsx')

target = 'Температура почвы (°C)'
X = df.drop(columns=[target, 'Дата и время'])
y = df[target]
```

```python
# ===== TRAIN / TEST SPLIT 70 / 30 =====
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, shuffle=False
)

# используем 1000 объектов обучающей выборки
X_train_1000 = X_train.iloc[:1000]
y_train_1000 = y_train.iloc[:1000]
```

```python
# ===== ЛИНЕЙНАЯ РЕГРЕССИЯ =====
model = LinearRegression()
model.fit(X_train_1000, y_train_1000)
```

```python
# ===== ОЦЕНКА КАЧЕСТВА =====
y_pred_train = model.predict(X_train_1000)
y_pred_test = model.predict(X_test)

def metrics(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    return mse, rmse, mae, r2

print('TRAIN (1000): MSE, RMSE, MAE, R2 =', metrics(y_train_1000, y_pred_train))
print('TEST:        MSE, RMSE, MAE, R2 =', metrics(y_test, y_pred_test))
```

```python
# ===== ГРАФИК: РЕАЛЬНЫЕ ЗНАЧЕНИЯ vs ПРЕДСКАЗАНИЯ =====
dates_test = df.loc[X_test.index, 'Дата и время']

plt.figure(figsize=(14, 5))
plt.plot(dates_test, y_test.values, label='Реальные значения', linewidth=2)
plt.plot(dates_test, y_pred_test, label='Предсказания модели', linewidth=2)
plt.xlabel('Дата и время')
plt.ylabel('Температура почвы (°C)')
plt.title('Реальные значения и предсказания линейной регрессии')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
